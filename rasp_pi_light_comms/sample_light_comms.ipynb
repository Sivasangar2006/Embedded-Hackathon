```python
import requests
import time

PI_IP="10.197.62.105"   # change this
PORT=8000

BASE_URL=f"http://{PI_IP}:{PORT}"

def led_on(led_id:int):
    url=f"{BASE_URL}/led/{led_id}/on"
    r=requests.post(url,timeout=2)
    print(r.json())

def led_fade_off(led_id:int,duration:float):
    url=f"{BASE_URL}/led/{led_id}/off"
    params={"duration":duration}
    r=requests.post(url,params=params,timeout=2)
    print(r.json())

# ---------------- EXAMPLE USAGE ---------------- #

# Person enters room 1
led_on(1)

# Simulate person staying inside for 12 seconds
time.sleep(12)

# Person exits → fade over 6 seconds
led_fade_off(1,3)

# Person re-enters before fade completes
time.sleep(2)
led_on(1)

```

```python
import requests
import time

PI_IP="10.197.62.105"
PORT=8000

BASE_URL=f"http://{PI_IP}:{PORT}"

def post(path,params=None):
    url=f"{BASE_URL}{path}"
    r=requests.post(url,params=params,timeout=2)
    try:
        print(f"[POST] {path} ->",r.json())
    except Exception:
        print(f"[POST] {path} ->",r.text)

def start_robbery(glow:int,flicker:int):
    post("/robbery/start",{"glow":glow,"flicker":flicker})

def stop_robbery():
    post("/robbery/stop")

# ---------------- TEST FLOW ---------------- #

if __name__=="__main__":
    print("Starting ROBBERY MODE")
    print("Glow LED 1, Flicker LED 2")

    start_robbery(glow=2,flicker=1)

    print("Robbery mode running for 8 seconds...")
    time.sleep(8)

    print("STOPPING robbery mode immediately")
    stop_robbery()

    print("Test complete")
```

```python
import cv2

STREAM_URL = "http://172.18.230.94:8080/video"

cap = cv2.VideoCapture(STREAM_URL)

if not cap.isOpened():
    print("❌ Cannot open IP Webcam stream")
    exit()

print("✅ Stream opened")

while True:
    ret, frame = cap.read()
    if not ret:
        print("❌ Frame grab failed")
        continue

    cv2.imshow("IP Webcam", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
